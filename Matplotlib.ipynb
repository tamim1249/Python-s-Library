```python
#matplotlib Ploting Function
#plt.plot(x,y,color="Color Name".Line style='Line_Style",Linewidth=value,marker="marker symbol",label="label name"
#plt.xlabel("text")
#plt.ylabel("text")
month=[1,2,3,4]
sales=[1000,1500,1200,1800]
plt.plot(month,sales,color="blue",linestyle=":",linewidth=2,marker='o',label="2025 sales data")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly sales data Report")
plt.legend(loc="best",fontsize=12)
plt.grid(color="gray",linestyle=':',linewidth=1)
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()
#plt.legend(loc='lower right', fontsize=10, shadow=True, title='Data Key', title_fontsize=12)
```

```python
#Bar Chart
#plt.bar(x,height,color='color name',width-value,label='label name')
product=['A','B','C','D']
sales=[1000,1500,800,1200]
plt.bar(product,sales,color='green',width=0.5,label='Sales 2025')
plt.xlabel("Product")
plt.ylabel('Sales')
plt.title('product sales Comparison')
plt.legend()
plt.show()
```

```python
import numpy as np

product = ['A', 'B', 'C', 'D']
sales_2024 = [900, 1200, 1000, 1100]
sales_2025 = [1000, 1500, 800, 1200]

x = np.arange(len(product))  # [0,1,2,3]

plt.bar(x - 0.2, sales_2024, width=0.4, label='2024', color='blue',edgecolor='black')
plt.bar(x + 0.2, sales_2025, width=0.4, label='2025', color='orange',edgecolor='black')

plt.xticks(x, product)
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Yearly Product Sales Comparison")
plt.legend()
plt.show()
```

```python
#plt.pie(values,labels=label_list,colors=color_list,autopct='%1.1f%%')
regions=['North','South','East','West']
revenue=[3000,2000,1500,1000]
plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors=['red','skyblue','lightgreen','coral'])
plt.title("Revenue Contribution By Region")
plt.show()
```

```python
#multiple pie chart
import matplotlib.pyplot as plt

regions = ['North', 'South', 'East', 'West']
revenue = [3000, 2000, 1500, 1000]

plt.subplot(1, 2, 1)
plt.pie(revenue, labels=regions, autopct='%1.1f%%')
plt.title("Revenue 2025")

plt.subplot(1, 2, 2)
plt.pie([2500, 2200, 1800, 900], labels=regions, autopct='%1.1f%%')
plt.title("Revenue 2024")

plt.tight_layout()
plt.show()

```

```python
#plt.scatter(x,y,color='color_name',marker='marker_style',label-'label_name')
hours_studies=[1,2,3,4,5,6,7,8]
exam_scores=[50,58,60,65,70,75,80,85]
plt.scatter(hours_studies,exam_scores,color='green',marker='o',label='Students Data')
plt.xlabel('Hours studies')
plt.ylabel('Exam_score')
plt.title("Relationship Between study Time and exam Score")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#multiple scatter
plt.scatter([1,2,3],[50,55,70],color='blue',label='Class A')
plt.scatter([1,2,3],[44,50,42],color='red',label='Class B')
plt.xlabel('Hours studies')
plt.ylabel('Exam_score')
plt.title("Comparison of two classes")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#plt.subplot(nrows,ncols,indes)
x=[1,2,3,4]
y=[10,20,15,25]
plt.subplot(1,2,1)#1st row,2nd column,1st subplot
plt.plot(x,y)
plt.title("Line Chart")

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title("Bar Chart")

plt.tight_layout()
plt.show()
```

```python

#fig,ax=plt.subplots(nrows,ncols,figsize=(width,height))

fig,ax=plt.subplots(1,2,figsize=(10,5))
x=[1,2,3,4]
y=[10,20,15,25]

ax[0].plot(x,y,color='black')
ax[0].set_title("Line Chart")

ax[1].bar(x,y,color='red')
ax[1].set_title("Bar Chart")

fig.suptitle("Comparison of Line and Bar Charts")
plt.tight_layout()
plt.show()
```
